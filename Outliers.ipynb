```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("train.csv")
garage = data["GarageArea"]
saleprice = data["SalePrice"]
```

```python
plt.scatter(garage, saleprice, color="blue")
```

```python
lowerbound = 0.1
upperbound = 0.95
outlier = garage.quantile([lowerbound, upperbound])
outlier
```

```python
final = ((garage>outlier.loc[lowerbound]) & (garage<outlier.loc[upperbound]))
result = garage[final]
```
